```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import array
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import math
import statsmodels.api as sm 
import datetime
import itertools
import logging
logging.getLogger('fbprophet').setLevel(logging.ERROR)
import warnings
from datetime import date
from imblearn.over_sampling import SMOTE
warnings.filterwarnings("ignore")

import fbprophet
from fbprophet import Prophet
from pandas import to_datetime
from scipy.stats import boxcox
from scipy.special import inv_boxcox
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error
```

```
df_train=pd.read_excel('data_abonents.xls',header=0)
```

```
df_train["Date"]=to_datetime(df_train["Date"])
data=df_train["Date"]
labels=df_train["series1"]
```

```
reg = GradientBoostingRegressor(random_state=1234)
reg.fit(data_train, labels_train)
```

```
from sklearn.model_selection import GridSearchCV
```

```
parametrs={
"loss":["ls", "lad" , "‘huber", "quantile"],
"n_estimators":[100,300,500,1000],
"learning_rate":[0.1, 0.3, 0.5]}
```

```
reg = GradientBoostingRegressor()
grid_search=GridSearchCV(reg,parametrs)
grid_search.fit(data_train, labels_train)
```

```
grid_search.best_params_
```

```
bestscroeXGboost=grid_search.best_score_
bestmodelXGboost=grid_search.best_estimator_
bestmodelXGboostPredict=bestmodelXGboost.predict(data_test)
```

```
parametrs={
"n_neighbors":[5,6,10],
"leaf_size":[100,40,30,59,200,1000],
"weights":["uniform","distance"],
"algorithm":["auto","ball_tree","kd_tree","brute"]}

neigh = KNeighborsRegressor()
neigh.fit(data_train, labels_train)
grid_searchKneigh=GridSearchCV(neigh,parametrs)
grid_searchKneigh.fit(data_train, labels_train)
```

```
grid_searchKneigh.best_params_
```

```
bestscroeKneigh =grid_searchKneigh.best_score_
bestmodelKneigh=grid_searchKneigh.best_estimator_
bestmodelKneighPredict=bestmodelKneigh.predict(data_test)
```

```
def pltPredictModel(modelpredic,bestscoremodel=0):

  plt.figure(figsize=[12,6])
  plt.plot(modelpredic[:200], label="predict")
  plt.plot(labels_test[:200],label="real_data")
  plt.legend()
  plt.show()
  print(f"bestscore by mae- {bestscoremodel}")
```

```
pltPredictModel(bestmodelXGboostPredict,bestscroeXGboost)
```

```

pltPredictModel(bestmodelKneighPredict,bestscroeKneigh)
```
